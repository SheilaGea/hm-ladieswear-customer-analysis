```python
# For my project, not all columns are necessary.
# I’ll keep product ID, name, type, group, department, index, garment group, and description.

articles_clean = articles[[
    "article_id", "product_code", "prod_name", "product_type_no", "product_type_name", "product_group_name",
    "colour_group_code", "colour_group_name", "department_name", "index_code", "index_name", "index_group_no", 
    "index_group_name", "section_no", "section_name", "garment_group_no", "garment_group_name"
]]
```

```python
# keep only Ladieswear articles 

ladieswear_articles = articles_clean[articles_clean["index_name"] == "Ladieswear"]
```

```python
import pandas as pd
```

```python
transactions = pd.read_csv("transactions_train.csv", parse_dates=["t_dat"])
```

```python
transactions
```

```python
# Check for duplicates 
transactions.duplicated().sum()
```

```python
# Check for missing values
transactions.isnull().sum()
```

```python
transactions[transactions.duplicated()].head()
```

```python
transactions_grouped = transactions.groupby(
    ["t_dat", "customer_id", "article_id", "price", "sales_channel_id"]).size().reset_index(name="quantity")
```

```python
transactions_grouped
```

```python
transactions_grouped.duplicated().sum()
```

```python
# Join tables and keep only articles that actually had transactions

ladieswear_full = ladieswear_articles.merge(transactions_grouped, on="article_id", how="inner")
```

```python
ladieswear_full
```

```python
import seaborn as sns
sns.histplot(ladieswear_full["price"])
```

```python
print(ladieswear_full["price"].mean())
print(ladieswear_full["price"].max())
print(ladieswear_full["price"].min())
```

```python

# Checking the prices

# Find the row(s) with the highest normalized price

max_price = ladieswear_full["price"].max()

highest_price_articles = ladieswear_full.loc[
    ladieswear_full["price"] == max_price, 
    ["article_id", "prod_name", "product_type_name", "price"]
]

print(highest_price_articles)
```

```python
# Define the real max price in SEK

# On the H&M website, the highest price for a LEATHER COAT is SEK 3.999,00

real_max_price = 3999  

# Rescale normalized prices

ladieswear_full["price_sek"] = ladieswear_full["price"] * (real_max_price / ladieswear_full["price"].max())
```

```python
ladieswear_full[["price", "price_sek"]].max()
```

```python
# Add round for 2 decimal places

ladieswear_full["price_sek"] = ladieswear_full["price_sek"].round(2)
```

```python
ladieswear_full[["prod_name", "price_sek"]].head()
```

```python
import pandas as pd

customers = pd.read_csv("customers.csv")
```

```python
print(customers.shape)
```

```python
print(customers.info())
```

```python
customers
```

```python
customers.columns = customers.columns.str.lower()
```

```python
customers.head()
```

```python
duplicates = customers["customer_id"].duplicated().sum()
duplicates
```

```python
print(customers.isna().sum())
```

```python
#  Drop irrelevant columns 

customers = customers.drop(columns=["fn", "fashion_news_frequency"])
```

```python
customers
```

```python
customers["active"].value_counts(dropna=False)
```

```python
# Keep only rows that are active

customers_active = customers.dropna(subset=["active"])
customers_active = customers_active[customers_active["active"] == 1]
```

```python
print(customers_active["active"].value_counts())
```

```python
customers_active.head()
```

```python
# Drop postal_code from active customers

customers_active = customers_active.drop(columns=["postal_code"])
```

```python
customers_active.head()
```

```python
# Merge with ladieswear_full

# how="inner" : keep only transactions from active customers

hm_data = ladieswear_full.merge(customers_active, on="customer_id", how="inner")
```
